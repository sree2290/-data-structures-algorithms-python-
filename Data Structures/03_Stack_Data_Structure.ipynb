## Stack

It is a linear data structure that follows a particular order in which the operations are performed.

LIFO( Last In First Out ):This strategy states that the element that is inserted last will come out first. You can take a pile of plates kept on top of each other as a real-life example. The plate which we put last is on the top and since we remove the plate that is at the top, we can say that the plate that was put last comes out first.

### Basic Operations on Stack

In order to make manipulations in a stack, there are certain operations provided to us.


*   push() to insert an element into the stack
*   pop() to remove an element from the stack
*   top() Returns the top element of the stack.
*   isEmpty() returns true is stack is empty else false
*   size() returns the size of stack

    




 Time Complexity
* Operations --Complexity
* push() ------O(1)
* pop() -------O(1)
* isEmpty() ---O(1)
* size() ------O(1)

Implementation of Stack Using List: 

```python
class Stack:
    def __init__(self):              # creating stack
        self.container = []
    
    def push(self,element):          # push opertaion to add elements at top
        self.container.append(element)

    def pop(self):                   # pop operation to remove elements from top
        self.container.pop()

    def print_stack(self):           # printing elements in stack
        print(self.container)
 
    def get_length(self):            # to get length of the stack
        return len(self.container)
    
    def peek(self):                  # To get the top element from stack
        return self.container[-1]

    def is_empty(self):              # To check whether the stack is empty or Not
        if len(self.container) == 0:
            return True
        else:
            return False
```

```python
s = Stack()
```

```python
s.push(8)
s.push(90)
s.push(23)
s.push(69)
```

```python
s.print_stack()
```

```python
s.pop()
```

```python
s.print_stack()
```

```python
s.get_length()
```

```python
s.peek()
```

```python
s.is_empty()
```

Implement Stack class using a deque


```python
from collections import deque
 
class Stack_deque:
    def __init__(self):              # creating stack
        self.container = deque()
    
    def push(self,element):          # push opertaion to add elements at top
        self.container.append(element)

    def pop(self):                   # pop operation to remove elements from top
        return self.container.pop()

    def print_stack(self):           # printing elements in stack
        print(self.container)
 
    def get_length(self):            # to get length of the stack
        return len(self.container)
    
    def peek(self):                  # To get the top element from stack
        return self.container[-1]

    def is_empty(self):              # To check whether the stack is empty or Not
        if len(self.container) == 0:
            return True
        else:
            return False
```

```python
s_d = Stack_deque()
```

```python
s_d.push(34)
s_d.push(12)
s_d.push(5)
s_d.push(78)
s_d.push(100)
s_d.push(798)

```

```python
s_d.print_stack()
```

```python
s_d.pop()
```

```python
s_d.print_stack()
```

```python
s_d.get_length()
```

```python
s_d.is_empty()
```

Implement Stack class using a LIFO Queue


```python
from queue import LifoQueue

 
class Stack_Queue:
    def __init__(self):              # creating stack  with the size of 6
        self.container = LifoQueue(maxsize=6)
    
    def push(self,element):          # put opertaion to add elements at top
        return self.container.put(element)

    def get(self):                   # get operation to remove elements from top
        return self.container.get()

    def get_length(self):            # to get length of the stack
        return self.container.qsize()
    
    def peek(self):                  # To get the top element from stack
        return self.container[-1]

    def is_full(self):
        return self.container.full()
```

```python
st_q = Stack_Queue()
```

```python
st_q.push(89)
st_q.push(23)
st_q.push(2)
st_q.push(59)
st_q.push(69)

```

```python
st_q.get()
```

```python
st_q.is_full()
```

```python
st_q.push(100)
st_q.push(890)
```

```python
st_q.is_full()  # Now stack became full with size of 6
```
